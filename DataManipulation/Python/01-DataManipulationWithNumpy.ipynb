### Operadores entre arrays

```python
array22 = np.arange(0,15,1)
array23 = np.arange(1,30,2)

subtracao = array22 - array23
soma = array22 + array23
divisao = array22 / array23
multiplicacao = array22 * array23

print(' v1', array22, '\n', 'v2', array23, '\n\n', 'subtracao (v1 - v2) = ', subtracao, '\n\n', 'soma (v1 + v2) = ', soma, '\n\n', 'divisao (v1 / v2) = ', divisao, '\n\n', 'multiplicacao (v1 * v2) = ', multiplicacao)
```

```python
array24 = np.array([True, False, True, False])
array25 = np.array([False, False, True, True])
print(' v1:         ', array24, '\n', 'v2:         ', array25, '\n\n', 'logical_and:', np.logical_and(array24, array25), '\n\n', 'logical_not:', np.logical_not(array24, array25), '\n\n', 'logical_or: ', np.logical_or(array24, array25), '\n\n', 'logical_xor:', np.logical_xor(array24, array25))
```

```python
#fatiamento de matrizes

array28 = np.array([[23,24,55,65], [20,14,30,80],[88,45,11,73]]);
print('matriz original: ', '\n', array28, '\n\n', 'shape (original):', array28.shape, '\n\n', 'matriz com slice (fatiada)', '\n', array28[:2,1:3], '\n\n', 'shape (fatiada):', (array28[:2,1:3]).shape)
```

```python
#atribuindo um valor a um elemento da Matriz

array29 = np.array([[10,11,55], [12,13,30],[23,78,14]]);
print('matriz original: ', '\n', array29, '\n\n')
array29[1,2] = 5555

print('matriz original com valor da posição [1,2] alterado: ', '\n', array29, '\n')

print('retornando o valor da linha [2] da matriz original:', '\n', array29[2], '\n')

print('retornando uma array de elementos de uma coluna específica da matriz original: ', '\n', array29[:,1], '\n')

print('retornando uma matriz de elementos de uma coluna específica da matriz original: ', '\n', array29[:,1:2], '\n\n', 'shape: ', (array29[:,1:2]).shape)
```

```python
array30 = np.arange(0,50,1) * 2; array30

#dentro e um vetor, quero os elementos entre as posições array.min() até (que seja <) o array30.max(),
#saltando de 2 em 2 (step).

#array30[array30.min():array30.max():2]

print(' array original: ', '\n', array30, '\n')
print(' dentro do array original, quero os elementos entre as posições array.min(menor valor do vetor) até (<)')

print(' o array.max(maior valor do vetor), saltando de 4 em 4 posições (step).')
print('', array30[array30.min():array30.max():4], '\n')

print(' retorna todos os valores até o elemento de indice 5 (exclusivo)', '\n', array30[:5], '\n')
print(' retorna todos os valores a partir do elemento de indice 5 (inclusivo)', '\n', array30[5:], '\n')
print(' dentro do array original, quero todos os elementos saltando de 10 em 10 posições (step).', '\n', array30[::10], '\n')

array30[:3] = 1
print(' atualizar os valores do arrays para 1 dos indices < 3 (exclusivo)', '\n', array30)
```

### Transposição e Reshaping de Arrays

```python
#Matriz transposta (pivotiamento) -> o que é coluna passa a ser linha, o que é linha passa a ser coluna

array34 = np.array([[1,3,80,90],[8,95,12,4]]); array34
print(' matriz original: ', '\n', array34, '\n')
print(' matriz transposta da matriz original (array.T ou numpy.transpose(array)): ', '\n', array34.T, '\n')
```

### Broadcasting

```python
# [x1, x2] -> lista, [[x1, x2]] -> array, [[x1, x2],[y1, y2]] -> matriz

array35 = np.array([[50, 10, 55, 89]]); array35
print(' array original: ', '\n', array35, '\n')
print(' array transposta do array original (array.T ou numpy.transpose(array)): ', '\n', array35.T, '\n')

# Broadcasting = replicação do array
print(' broadcasting = replicação do array (array oiginal + transposta do array original): ', '\n', array35 + array35.T, '\n')
```

### Transformando Matrizes em Arrays
#### numpy.ravel() -> transforma um objeto multidimensional (matriz) em unidimensional (array)

```python
#numpy.ravel() -> transforma um objeto multidimensional (matriz) em unidimensional (array)

array36 = np.array([np.arange(1,100,20), np.arange(1,200,40), np.arange(1,300,60)])

print(' array original: ', '\n', array36, '\n')
print(' transformando um objeto multidimensional (matriz original) em um unidimensional (array): ', '\n', array36.ravel(), '\n')
print(' array transposta do array original (array.T ou numpy.transpose(array)): ', '\n', array36.T, '\n')
print(' transformando um objeto multidimensional (matriz original TRANSPOSTA) em um unidimensional (array): ', '\n', array36.T.ravel(), '\n')
```

### Ordenando o Array

```python
array39 = np.array([10,55,65,1,78,102,3])
print(' array original (desordenado): ', '\n', array39, '\n')
array39.sort()
print(' array original (ordenado): ', '\n', array39, '\n', 'shape do array:', array39.shape, '\n')

array40 = np.array([[10,55,65,1], [65,2,36,20], [115,20,55,40]])
array41 = np.array([[10,55,65,1], [65,2,36,20], [115,20,55,40]])
print(' array´s da matriz original (desordenado): ', '\n', array40, '\n')
array40.sort(axis=1) #ordenado por linha
array41.sort(axis=0) #ordenado por coluna
print(' array´s da matriz original (ordenado por linha):', '\n', array40, '\n', 'shape da matriz:', array40.shape, '\n')
print(' array´s da matriz original (ordenado por coluna):', '\n', array41, '\n', 'shape da matriz:', array41.shape, '\n')
```
